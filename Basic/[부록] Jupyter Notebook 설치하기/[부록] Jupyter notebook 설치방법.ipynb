## Jupyter Notebook 설치방법

### Jupyter Notebook 이란? ###

파이썬을 손쉽게 실행하고, 다룰 수 있도록 도와주는 툴입니다.  
Pycharm과 더불어 Python 사용자들이 가장 많이 사용하는 IDE (통합 개발환경) 중 하나입니다.  
Jupyter Notebook은 PIP를 통해 손쉽게 설치가 가능합니다.

### PIP란?

파이썬의 패키지 매니저로서, 윈도우의 '프로그램 추가/제거'와 비슷하게  
파이썬에서 사용되는 외부 라이브러리(플러그인)들을 관리해주는 역할을 하는 프로그램입니다.

**이 프로그램은 따로 설치할 필요 없이 파이썬을 설치할 때 자동으로 설치됩니다.**

### 설치 과정  
먼저, "명령 프롬프트"를 실행합니다.

`pip`를 콘솔창에 입력하여 pip가 PC에 설치되어있는지 확인해봅니다.

위와 같이 나온다면 Jupyter Notebook을 설치할 준비가 완료된 것입니다.

### 'pip'은(는) 내부 또는 외부 명령, 실행할 수 있는 프로그램, 또는 배치 파일이 아닙니다. 라고 나오는 경우 

bin 폴더가 환경변수에 등록되어있지 않아서 발생한 문제입니다.  
아래 링크를 참조하여 pip가 정상적으로 실행될 수 있게 만들어주세요.  
https://somjang.tistory.com/249 

### pip를 설치합니다.

아래 명령어를 콘솔에 입력해주세요.  
`pip install jupyter`

빨간 글씨가 뜨지 않고, 성공적으로 설치가 완료되었습니다.  
콘솔 창에 아래 명령어를 입력하여 Jupyter Notebook을 실행해주세요.  

`Jupyter Notebook `
