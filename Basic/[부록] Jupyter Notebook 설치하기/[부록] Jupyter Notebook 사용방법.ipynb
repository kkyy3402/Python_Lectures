### Jupyter Notebook 좀더 알아보기

위와 같이 각 행의 좌측에 표시되어 있는 부분은, 현재 선택된 행입니다.  
선택된 행에서 Shift + Enter를 했을 때 파이썬 명령어가 실행됩니다.  

*누가봐도 선택된 행인 것을 알 수 있는데, 여러분을 제가 너무 과소평가 했나요*

왼쪽에 뜬 이렇게 초록색인 경우가 있습니다.  
초록색 : 편집중인 행을 의미합니다.  ('Edit' 모드라고 합니다.)  
파란색 : 선택만 되고, 편집중이 아닌 행을 의미합니다. ('Command' 모드라고 합니다.)

각 행에서 Esc를 누르면 파란색, Enter를 눌러 편집을 시작하면 초록색으로 변경되는 것을 알 수 잇습니다.

그리고, 위 사진을 보면 ln [2] 라고 되어있는데,  
ln은 라인을 의미하고, [2]는 해당 파일에서 몇번째 실행인지를 나타냅니다.

## 셀 다루기

하나의 행에 모든 코드를 실행할 수 있지만,  
너무 복잡해지고, 가독성이 많이 떨어지겠죠,  
그래서 행을 삽입하고, 삭제, 수정하는 방법에 대해 다뤄보겠습니다.

**Jupyter Notebook에서는 각 행을 '셀'이라고 부릅니다. 따라서, 지금부터는 행을 '셀' 이라고 부르겠습니다.**

### 셀의 삽입
셀의 삽입은 메뉴의 Insert 항목을 통해 실행할 수 있습니다. 

이 작업도 물론 단축키를 통해 실행 가능합니다.  
`A ('Above'의 약자)` : 선택된 행의 위에 셀을 삽입합니다.  
`B ('Below'의 약자)` : 선택된 행의 아래에 셀을 삽입합니다.



### 행의 삭제 ###   
셀의 삭제는 셀을 선택한 후, 메뉴의 `Delete Cells`를 선택하여 삭제할 수 있습니다.  
단축키 : `d` 를 두번 누릅니다.

### 셀의 수정
셀의 수정은 간단합니다.  
Command Mode에서 각 셀을 선택하고,  엔터를 누른 후 메모장처럼 텍스트를 수정하시면 됩니다.




### 노트북 저장하기 ###  

노트북을 저장하는 방법은 간단합니다.  
일반적인 메모장에서 저장 단축키인 `Ctrl + S` 단축키를 통해서 하는 방법과,  
[File] -> [Save and CheckPoint]를 클릭하여 저장하는 방법이 있습니다.
