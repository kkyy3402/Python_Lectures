### 우리는 왜 컴퓨터를 사용할까? ###

이유를 알면 좀더 재미있게 배울 수 있을 것 같아,  
누군가에게는 쓰잘대기 없다고 느껴질 수 있는 TMI를 적어보고자 합니다. 

우리가 프로그래밍을 배우기 전에,  
먼저 우리가 컴퓨터를 왜 사용할까요?

그 이유는, 컴퓨터는 우리보다 단순 연산작업을 훨씬 빨리(라고 말해도 비교가 안될정도로 빠릅니다) 수행하기 때문입니다. 

이러한 엄청한 수행속도는 우리로 하여금 엑셀, 게임, 그래픽 작업 등을 가능하게 하여 생산성의 향상에 기여합니다.

결국 프로그래밍을 배운다는 것은  
**내가 가진 문제의 해결을 위해 이러한 컴퓨터의 엄청난 수행속도를 빌리는 방법을 배우는 것**이라 생각하시면 어떨까 싶습니다.

### 왜 파이썬인가? ###

요새 파이썬이 그렇게 좋다고 난리입니다.  
왜 배울까요? 장단점을 한번 써보겠습니다.

장점
- **배우기 쉽습니다** : 문법이 사람의 언어와 상당히 비슷한면이 많습니다. 그리고 쉽습니다. 이 점이 가장 큰 장점이라고 생각합니다.
- **라이브러리가 풍부하다** : 머신러닝 등에 사용할 수 있는 매우 방대한 라이브러리 생태계가 형성되어있습니다. 

단점 
- **속도가 느리다** : 파이썬은 대표적인 '인터프리터'언어인데, 이 언어의 특징은 느리다는 특징이 있습니다. 하지만, 현재 우리가 하려는 입문단계의 프로그래밍 수준에서는 실행에 전혀 지장이 없는 수준입니다. 
- 모바일 프로그래밍, 임베디드 프로그래밍을 하기는 어렵습니다. 현재 기준에서 안드로이드, 아이폰 개발을 위해서는 다른 언어를 배우셔야합니다.

입문 단계의 전달하고자 하는 내용이어서,  
기술하지 장단점들이 있지만  
배우기 쉽다는 장점 하나가 모든 단점을 커버할 만큼 강력합니다.  

여러가지 프로그래밍 언어들이 있습니다.  
각 언어별로 조금씩 다른 면이 있지만, 대부분 비슷한 내용을 다루고 있어,  
한 언어를 잘 공부해두면 몇몇 경우를 제외하고는 다른 언어에 적응하는데 크게 오래 걸리지 않습니다.

따라서, 이번 기회에 파이썬 공부를 꼭 완주하시길 바라겠습니다.
