### 라이브러리란?
Library는 도서관이라는 뜻이죠.  
전세계에 있는 파이썬을 사용하는 사람들이 만들어놓은 **유용한 도구모음**으로 생각해주시면 될 것 같습니다.  
파이썬에서는 pip라는 도구를 통해 손쉽게 라이브러리를 설치하고 삭제할 수 있습니다.  
윈도우의 '프로그램 추가/제거' 혹은 리눅스의 'apt'와 비슷한 역할을 한다고 생각해주시면 됩니다.

이 라이브러리를 통해 크롤링, 데이터 시각화, GUI 프로그래밍, 네트워킹 등 다양한 작업들을 파이썬을 통해 할 수 있습니다.

### pip를 통해 라이브러리 설치해보기

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)    # Reproducible random state

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
